```python
import pandas as pd, matplotlib.pyplot as plt, geopandas as gpd, contextily as ctx, numpy as np
```

```python
data = "CABA_radios.shp"
radio = gpd.read_file(data)
```

```python
radio["RADIO_REF_ID"] = radio.index
radio
```

```python
radios = radio[["RADIO_REF_ID","RADIO","DPTO","IDDPTO","geometry"]]
```

```python
radios
```

```python
data2 = "censo_base.csv"
censo_base = pd.read_csv(data2)
```

```python
censo_base
```

Mergear para mapear

```python
mergear = censo_base[["RADIO_REF_ID","TELEFONO"]]
mergear
```

```python
tabla1 = (mergear.groupby(["RADIO_REF_ID","TELEFONO"])
          .size()
          .to_frame()
)

tabla1
```

```python
radios_inner = pd.merge(tabla1,radios, on="RADIO_REF_ID", how="inner")
radios_inner
```

```python
radios_inner = pd.merge(mergear, radios, on="RADIO_REF_ID", how="inner")
radios_inner
```
