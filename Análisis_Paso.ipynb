```python
import pandas as pd, matplotlib.pyplot as plt, geopandas as gpd, contextily as ctx, numpy as np
```

```python
censo = gpd.read_file("CABA.shp")

censo
```

```python
df = pd.read_csv("paso_x_partido.csv")

(df[["circuito","pp2"]]
    .groupby(["circuito"])
    .agg(["mean","std"])
    .stack()
)
```

```python
cdg = censo[["circuito","departamen","geometry"]].sort_values(by="circuito").reset_index(drop=True)
cdg
```

```python
cdg.dtypes
```

```python
cdg = cdg.astype({"circuito":int})
cdg.dtypes
```

```python
abc = pd.read_csv("paso_x_partido.csv")

(df[["circuito","pp2"]]
    .groupby(["circuito"])
    .agg(["sum"])
)
```

```python
abc.dtypes
```

```python
informe_inner = pd.merge(cdg, abc, on="circuito", how="inner")
informe_inner
```

```python
informe_inner.groupby(["departamen"])[["pp1", "pp2", "pp3", "pp4", "nv", "geometry"]].sum().sort_values(by=["pp2"], ascending=False)
```
