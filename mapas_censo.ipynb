```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, geopandas as gpd, contextily as ctx
```

```python
censo = pd.read_csv(
    r'C:\Users\leand_5odr3v\Desktop\rocio python\Data met4op\simulacion\censo\censo_conhogar.csv',
    delimiter=',',
    header=0,
    names=None,
    index_col=0,
    usecols=None,
    dtype=None,
    skiprows=None,
    skipfooter=0,
    nrows=None,
    decimal='.', 
    quotechar='"',
)
censo
```

```python
censo.rename(columns={'MUNI':'COMUNA', 'H19D':'TELEFONO', 'P02':'SEXO', 'P03':'EDAD', 'P09':'NIVEL_EDU'}, inplace=True)
```

```python
censo
```

```python
edadxradio = (censo[["RADIO_REF_ID","EDAD"]]
    .groupby(["RADIO_REF_ID"])
    .agg(["mean", "std"])
)
edadxradio
```

```python
mayores16 = censo.loc[(censo["EDAD"] > 16), ["RADIO_REF_ID", "EDAD", "NIVEL_EDU"]]
mayores16
```

```python
nbixradio = (censo[["RADIO_REF_ID","ALGUNBI"]]
    .groupby(["RADIO_REF_ID"])
    .agg(["mean", "std"])
)
nbixradio
```

```python
eduxradio2 = (mayores16[["RADIO_REF_ID","NIVEL_EDU"]]
    .groupby(["RADIO_REF_ID"])
    .agg(["mean", "std"])
)
eduxradio2
```

```python
educacionxradio = (censo[["RADIO_REF_ID","NIVEL_EDU"]]
    .groupby(["RADIO_REF_ID"])
    .agg(["mean", "std"])
)
educacionxradio
```
