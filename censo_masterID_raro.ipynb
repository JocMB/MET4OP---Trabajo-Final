```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import numpy as np
```

```python
dpto = pd.read_csv("dpto.csv")
dpto
```

```python
prov = pd.read_csv("prov.csv")
prov
```

```python
prov_dpto_in = pd.merge(prov, dpto, on="PROV_REF_ID", how="inner")
prov_dpto_in
```

```python
frac = pd.read_csv("frac.csv")
frac
```

```python
frac_dp_in = pd.merge(frac, prov_dpto_in, on="DPTO_REF_ID", how="inner")
frac_dp_in
```

```python
radio = pd.read_csv("radio.csv")
radio
```

```python
radio_frac_in = pd.merge(radio, frac_dp_in, on="FRAC_REF_ID", how="inner")
```

```python
radio_frac_in.columns
```

```python
radio_frac_in["IDRADIO"] = radio_frac_in["IDRADIO"].astype(str)
radio_frac_in["IDRADIO"] = radio_frac_in["IDRADIO"].str.pad(width=2, side="left", fillchar="0")
radio_frac_in["IDRADIO"]
```

```python
radio_frac_in["IDDPTO"] = radio_frac_in["IDDPTO"].astype(str)
radio_frac_in["IDDPTO"] = radio_frac_in["IDDPTO"].str.pad(width=3, side="left", fillchar="0")
radio_frac_in["IDDPTO"]
```

```python
radio_frac_in["IDFRAC"] = radio_frac_in["IDFRAC"].astype(str)
radio_frac_in["IDFRAC"] = radio_frac_in["IDFRAC"].str.pad(width=2, side="left", fillchar="0")
radio_frac_in["IDFRAC"]
```

```python
radio_frac_in["IDPROV"] = radio_frac_in["IDPROV"].astype(str)
radio_frac_in["IDPROV"] = radio_frac_in["IDPROV"].str.pad(width=2, side="left", fillchar="0")
radio_frac_in["IDPROV"]
```

```python
MasterID = radio_frac_in["MasterID"] = radio_frac_in["IDPROV"] + radio_frac_in["IDDPTO"] + radio_frac_in["IDFRAC"] + radio_frac_in["IDRADIO"]
MasterID
```

```python
MasterID = MasterID.to_frame()
```

```python
MasterID = MasterID.rename(columns={0:"MasterID"})
MasterID
```
