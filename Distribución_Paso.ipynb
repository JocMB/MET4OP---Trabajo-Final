```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import numpy as np
```

```python
pasoxpartido = pd.read_csv(
    'paso_x_partido.csv',
    delimiter=',', 
    index_col=0
)
pasoxpartido
```

```python
agrupxpartido = pd.read_csv(
    'datos_agrup.csv',
    delimiter=',', 
    index_col=0
)
agrupxpartido
```

```python
#Las comunas están asociadas con un intervalo de código de circuitos. 
#Por lo tanto, es posible agrupar los circuitos por comuna y rescatar cuánto acumula por comuna nuestro candidato 
intervalo_comuna = (agrupxpartido[["NOMBRE_REGION","CODIGO_CIRCUITO"]]
    .groupby(["NOMBRE_REGION"])
    .agg(["min","max"])
)
intervalo_comuna
```

```python
#Agrupamos los votos de cada partido por circuito
votosxcircuito = (pasoxpartido[["circuito" , "pp2", "pp1" , "pp3" , "pp4" , "nv"]]
    .groupby(["circuito"])
    .agg(["sum"])
)
votosxcircuito
```

```python
#Agregamos un índice
votosxcircuito.reset_index(inplace=True)
votosxcircuito
```

```python
#Agregamos la columna "suma_votos" para conocer la cantidad de votantes por circuito
votosxcircuito['suma_votos'] = votosxcircuito.loc[0:167,["pp2","pp1",	"pp3", "pp4","nv"]].sum(axis=1)
votosxcircuito
```

```python
#Se suma cada variable para conocer: 1) la cantidad de votos que recibió cada partido y 2) los votos totales 
nueva_fila= votosxcircuito.sum()
nueva_fila
```

```python
censo = gpd.read_file('CABA.shp')
censo
```

```python
#Se limpia la tabla para conservar las variables requeridas
cdg = censo[["circuito","departamen","geometry"]].sort_values(by="circuito").reset_index(drop=True)
cdg
```

```python
cdg.dtypes
```

```python
#Se modifica el tipo de dato en la variable "circuito" para combinar tablas 
cdg = cdg.astype({"circuito":int})
cdg.dtypes
```

```python
#Para la tabla abc (votoxpartido -hay que unificar nombre de tablas-) se agrupa para conocer los datos de nuestro candidato
abc = pd.read_csv('paso_x_partido.csv')
(pasoxpartido[["circuito","pp2"]]
    .groupby(["circuito"])
    .agg(["sum"])
)
```

```python
#Se combinan las tablas cdg y abc/votosxpartido para posteriormente utilizar geopandas con la variable "geometry"
informe_inner = pd.merge(cdg, abc, on="circuito", how="inner")
informe_inner
```

```python
#Se busca conocer la distribución porcentual de votos para nuestro candidato por comuna
porcentaje_comuna = informe_inner[["departamen","pp2"]].groupby(["departamen"]).agg(["sum"])
porcentaje_comuna.apply(lambda x: x / float(x.sum())*100)
```

```python
#Se añade la columna "suma_votos" para posteriormente conocer la cantidad de votantes por comuna y realizar un mapeo de dicho dato 
informe_inner['suma_votos'] = informe_inner.loc[0:7639,["pp2","pp1","pp3","pp4","nv"]].sum(axis=1)
informe_inner
```

```python
#Se agrupa por "departamen" y "suma_votos" para conocer la cantidad de votantes por comuna 
agrupadoxcomuna = (informe_inner[["departamen","geometry","suma_votos"]]
    .groupby(["departamen"])
    .agg(["sum"]))
agrupadoxcomuna
```
