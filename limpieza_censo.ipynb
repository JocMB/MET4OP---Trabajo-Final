## LIMPIAR LOS DATA FRAMES PARA EL MERGE

### Vivivendas

```python
vivienda_radio = censo_vivienda[["RADIO_REF_ID"]]
vivienda_radio
```

```python
vivienda_radio.reset_index(inplace=True)
vivienda_radio
```

### Hogares

```python
select_censo_hogar = censo_hogar[["VIVIENDA_REF_ID", "H19D", "ALGUNBI"]]
select_censo_hogar
```

```python
censo_limpio = pd.merge(hogares_viviendas, select_censo_ind, on="HOGAR_REF_ID", how="inner")
censo_limpio
```
