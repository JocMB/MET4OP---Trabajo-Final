## LIMPIAR LOS DATA FRAMES PARA EL MERGE

### Población

```python
select_persona = censo_persona[["HOGAR_REF_ID", "P02", "P03", "EDADQUI", "P09", "CONDACT"]]
select_persona
```

```python
select_persona.reset_index(inplace=True)
```

```python
select_persona
```

### Vivivendas

```python
select_vivienda = censo_vivienda[["RADIO_REF_ID", "MUNI"]]
select_vivienda
```

```python
select_vivienda.reset_index(inplace=True)
select_vivienda
```

### Hogares

```python
select_hogar = censo_hogar[["VIVIENDA_REF_ID", "H19D", "INDHAC", "ALGUNBI", "PROP"]]
select_hogar
```

```python
select_hogar.reset_index(inplace=True)
select_hogar
```

## Merge

```python
hogares_viviendas = pd.merge(select_vivienda, select_hogar, on="VIVIENDA_REF_ID", how="inner")
hogares_viviendas
```

```python
censo_base = pd.merge(hogares_viviendas, select_persona, on="HOGAR_REF_ID", how="inner")
censo_base
```

```python
censo_base = censo_base[[ "PERSONA_REF_ID", "RADIO_REF_ID", "MUNI", "H19D", "P02", "P03", "EDADQUI", "P09" , "CONDACT", "ALGUNBI", "INDHAC" ]]
censo_base
```

```python
censo_base.rename(columns={'MUNI':'COMUNA', 'H19D':'TELEFONO', 'P02':'SEXO', 'P03':'EDAD', 'P09':'NIVEL_EDU'}, inplace=True)
```
