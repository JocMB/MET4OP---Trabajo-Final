```python
hogar_vivienda_personas_ = pd.read_csv("censo_madre.csv")
hogar_vivienda_personas_
```

```python
tabla_intermedia0 = pd.read_csv("intermedia.csv")
tabla_intermedia0
```

```python
tabla_intermedia0.drop(["Unnamed: 0"], axis=1)
tabla_intermedia = tabla_intermedia0[["RADIO", "RADIO_REF_ID",	"circuito",	"RADIO.CIRCUITO"]]
tabla_intermedia
```

```python
tabla_intermedia = tabla_intermedia.rename(columns={'circuito': 'circuito0'})
tabla_intermedia
```

```python
tabla_intermedia["circuito0"]= tabla_intermedia["circuito0"].astype(str)
```

```python

strp = tabla_intermedia["circuito0"].str.pad(width=4, side="left", fillchar="0")
strp = strp.to_frame()
```

```python
strp = strp.rename(columns={'circuito0': 'circuito'})
strp
```

```python
tabla_intermedia_c = pd.concat((tabla_intermedia, strp), axis=1)
```

```python
tabla_intermedia_c.drop('circuito0', axis=1, inplace=True)
```
